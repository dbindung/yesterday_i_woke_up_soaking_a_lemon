# 2. Загрузка модели Qwen/Qwen2-7B-Instruct-AWQ

```python
from vllm import LLM

# Инициализация модели
model = LLM(
"Qwen/Qwen2-7B-Instruct-AWQ",
tensor_parallel_size=1, # Увеличьте, если используете несколько GPU
)
```

# 3. Примеры запросов к модели

```python
from vllm import SamplingParams

# Настройка параметров генерации
sampling_params = SamplingParams(
temperature=0.7,
top_p=0.95,
max_tokens=200
)

prompt = "Напиши краткое описание библиотеки vLLM"
out = model.generate(prompt, sampling_params)
print(out[0].outputs[0].text)
```
